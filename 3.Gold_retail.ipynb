```
from pyspark.sql.window import Window
from pyspark.sql.functions import col, avg, when

# Load Silver table
silver_df = spark.read.table("silver_daily_sales")

# Define window spec using the correct column name
window_spec = Window.orderBy("date").rowsBetween(-6, 0)

# Apply transformations
gold_df = silver_df.withColumn(
    "rolling_avg_sales", avg("quantity").over(window_spec)
).withColumn(
    "low_stock_flag", when(col("stock") < 500, 1).otherwise(0)
).withColumn(
    "price_tier", when(col("price") < 1.10, "Low")
        .when((col("price") >= 1.10) & (col("price") <= 1.20), "Medium")
        .when(col("price") > 1.20, "High")
        .otherwise("Unknown")
)

# Display result
display(gold_df)
```

```
gold_df.write.format("delta").mode("overwrite").saveAsTable("gold_daily_sales")
```
